#**Control de Dosificación de Fertilizante Líquido**

Este ejercicio es ideal para introducir float, int y la concatenación de strings.

**Planteamiento:**

Un sistema de riego por goteo necesita calcular el total de fertilizante (TF). El operario debe ingresar las hectáreas (ht) y la dosis (ds) en litros por hectárea.

**Requerimientos:**
1. Solicitar las hectáreas totales (Tipo float). (ht)
2. Solicitar la dosis en litros por hectárea (Tipo float). (ds)
3. Calcular el total de fertilizante (Tipo float). (tf)
Fórmulas:

$$tf=ht*ds$$

**Prueba del programa:**

Si se ingresan 50 hectáreas y 20 litros/hectárea, el programa debe informar que el total de fertilizante es de 1000 litros.

**Análisis (EPS)**

Entrada (E):

ht → hectáreas (float)

ds → dosis (float)

Proceso (P):

Calcular:

tf=ht*ds

Salida (S):

Cantidad total de fertilizante (tf)

```python
# ENTRADA: Solicitar las hectáreas y la dosis
ht = float(input("Ingrese las hectáreas totales: "))
ds = float(input("Ingrese la dosis en litros por hectárea: "))

# PROCESO: calcular el total de fertilizante (tf)
tf = ht * ds

# SALIDA: informar el resultado
print("El total de fertilizante es", tf, "litros")
```

### Explicación Resumida del Código:

*   **`ht = float(input("..."))` y `ds = float(input("..."))`**
    *   Estas líneas solicitan al usuario las hectáreas totales (`ht`) y la dosis por hectárea (`ds`), convirtiendo ambos valores a números decimales (`float`).

*   **`tf = ht * ds`**
    *   Esta línea calcula el total de fertilizante (`tf`) multiplicando las hectáreas por la dosis.

*   **`print("El total de fertilizante es", tf, "litros")`**
    *   Esta línea muestra el resultado final del cálculo (`tf`) al usuario en un mensaje fácil de entender.

### Prueba de Escritorio

**Datos de Entrada:**
*   Hectáreas (`ht`) = 50
*   Dosis (`ds`) = 20 litros/hectárea

**Fórmula:**
$$tf = ht \times ds$$

**Reemplazando los valores en la fórmula:**
$$tf = 50 \times 20$$
$$tf = 1000$$

**Resultado:**
El total de fertilizante (`tf`) es 1000 litros.
