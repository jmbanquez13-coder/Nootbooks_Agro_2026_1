**Control de Rendimiento de Pulpa en Frutas**

Este ejercicio es ideal para aplicar variables tipo float y operaciones aritméticas en el cálculo de rendimiento en procesos agroindustriales.

**Planteamiento:**

Una planta procesadora necesita calcular el rendimiento de pulpa (rp). El operario debe ingresar el peso total de fruta (pt) y el peso de los residuos (rs) para determinar la cantidad de pulpa obtenida y su porcentaje de aprovechamiento.

**Requerimientos:**

1. Solicitar el peso total de la fruta en kg (Tipo float). (pt)
2. Solicitar el peso de los residuos en kg (Tipo float). (rs)
3. Calcular la cantidad de pulpa obtenida:

$$pl=pt−rs$$

4. Calcular el porcentaje de rendimiento:


$$rp=(pl/pt)∗100$$

**Prueba del programa:**

Si se ingresan 500 kg de fruta y 150 kg de residuos, el programa debe calcular:

Pulpa = 350 kg
Rendimiento = 70 %

#**Análisis (EPS)**

**Entrada (E):**

pt → peso total de la fruta (kg) (Tipo float)

rs → peso de los residuos (kg) (Tipo float)

**Proceso (P):**

Calcular la cantidad de pulpa obtenida:

pl=pt−rs

Calcular el porcentaje de rendimiento:

rp=(pl/pt)∗100

**Salida (S):**

pl → cantidad de pulpa obtenida (kg)

rp → rendimiento de pulpa (%)

```python
# ENTRADA: Solicitar el peso total y los residuos
pt = float(input("Ingrese el peso total de la fruta (kg): "))
rs = float(input("Ingrese el peso de los residuos (kg): "))

# PROCESO: calcular la pulpa (pl) y el rendimiento (rp)
pl = pt - rs
rp = (pl / pt) * 100

# SALIDA: informar resultados
print("La cantidad de pulpa es", pl, "kg")
print("El rendimiento es", round(rp, 2), "%")
```

## Explicación del Código

**1. `pt = float(input("Ingrese el peso total de la fruta (kg): "))`**

*   **`input("...")`**: Esta función solicita al usuario que ingrese un valor a través de la consola y lo devuelve como una cadena de texto (string).
*   **`float(...)`**: Esta función toma la cadena de texto ingresada por el usuario y la convierte a un número de punto flotante (decimal). Esto es importante porque los pesos pueden tener decimales.
*   **`pt = ...`**: El valor numérico resultante se asigna a la variable `pt`, que representa el peso total de la fruta.

## Prueba de Escritorio

**Datos de Entrada:**
*   `pt` (peso total de la fruta) = 500 kg
*   `rs` (peso de los residuos) = 150 kg

**1. Calcular la cantidad de pulpa obtenida (`pl`):**

   Fórmula: `pl = pt - rs`

   Reemplazando valores:
   `pl = 500 - 150`
   `pl = 350 kg`

**2. Calcular el porcentaje de rendimiento (`rp`):**

   Fórmula: `rp = (pl / pt) * 100`

   Reemplazando valores:
   `rp = (350 / 500) * 100`
   `rp = 0.7 * 100`
   `rp = 70 %`

**Resultados:**
*   Pulpa = 350 kg
*   Rendimiento = 70 %
