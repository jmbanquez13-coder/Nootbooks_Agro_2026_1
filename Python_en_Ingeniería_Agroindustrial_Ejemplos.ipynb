## ¿Qué es Python?

Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Es famoso por ser:

* **Legible:** Es fácil de leer y escribir, lo que permite que ingenieros que no son programadores expertos lo aprendan rápido.
* **Versátil:** Sirve para analizar datos, controlar maquinaria, crear apps o desarrollar inteligencia artificial.
* **Comunidad gigante:** Existen miles de "librerías" (herramientas prefabricadas) gratuitas que resuelven problemas específicos de ingeniería.

## 3 Ejemplos sencillos de Python en Agroindustria

Aquí tienes tres scripts básicos que ilustran cómo podrías usarlo en tu día a día:

### 1. Estimación de rendimiento de cosecha

Este código ayuda a proyectar cuántas toneladas se obtendrán basadas en una muestra pequeña.

```
# Datos de entrada
area_total_ha = 50  # Hectáreas
plantas_por_m2 = 8
peso_promedio_fruto_gr = 150

# Cálculo (1 ha = 10,000 m2)
rendimiento_total_kg = (area_total_ha * 10000) * plantas_por_m2 * (peso_promedio_fruto_gr / 1000)
toneladas = rendimiento_total_kg / 1000

print(f"El rendimiento estimado para {area_total_ha} hectáreas es de {toneladas} toneladas.")
```

### 2. Alerta de condiciones climáticas (Sensores)

Simula el monitoreo de un invernadero para decidir si se activa el riego.

```
humedad_suelo = 15  # Porcentaje obtenido de un sensor

if humedad_suelo < 20:
    print("ALERTA: Humedad baja. Activando sistema de riego automático.")
elif humedad_suelo > 80:
    print("ALERTA: Humedad excesiva. Revisar drenaje.")
else:
    print("Condiciones óptimas. No se requiere acción.")
```

### 3. Balance de Mezcla (Cuadrado de Pearson simple)

Útil para formular alimentos balanceados para animales buscando un porcentaje específico de proteína.

```
# Queremos una mezcla con 15% de proteína
proteina_maiz = 9
proteina_soya = 44
objetivo = 15

# Cálculo de partes
partes_soya = abs(objetivo - proteina_maiz)
partes_maiz = abs(objetivo - proteina_soya)
total_partes = partes_soya + partes_maiz

porcentaje_maiz = (partes_maiz / total_partes) * 100

print(f"Para obtener {objetivo}% de proteína, la mezcla debe tener {porcentaje_maiz:.2f}% de maíz.")
```
