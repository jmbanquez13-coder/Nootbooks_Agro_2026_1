#**Control de Capacidad en Silos de Granos**

Este ejercicio es ideal para introducir float, int y la concatenación de strings.

**Planteamiento:**

Una planta agroindustrial necesita calcular la capacidad de almacenamiento de un silo (CS). El operario debe ingresar el radio (r) y la altura (h) del silo en metros.

**Requerimientos:**
1. Solicitar el radio del silo en metros (Tipo float). (r)
2. Solicitar la altura del silo en metros (Tipo float). (h)
3. Calcular el volumen del silo (Tipo float). (V)
4. Calcular la capacidad en toneladas considerando que la densidad del maíz es 0.75 ton/m³. (cs)


  $$V=π*r^2*h$$

  $$cs=V*0.75$$


**Prueba del programa:**

Si se ingresan 3 m de radio y 10 m de altura, el programa debe informar que la capacidad del silo es de aproximadamente 212.06 toneladas.

#**Análisis (E-P-S)**

Entrada (E):

r → radio del silo (float)

h → altura del silo (float)

Proceso (P):

Calcular el volumen:


V=π*r^2*h

Calcular la capacidad:

cs=V×0.75


Salida (S):

Capacidad del silo en toneladas (cs)

```python
import math

# ENTRADA: Solicitar el radio y la altura del silo
r = float(input("Ingrese el radio del silo en metros: "))
h = float(input("Ingrese la altura del silo en metros: "))

# PROCESO: calcular el volumen (V) y la capacidad (cs)
V = math.pi * (r ** 2) * h
cs = V * 0.75

# SALIDA: informar la capacidad del silo
print("La capacidad del silo es de", round(cs, 2), "toneladas")
```

El código realiza los siguientes pasos:

1.  **`import math`**: Importa la librería `math` para usar `math.pi`.

2.  **`r = float(input(...))`**: Solicita al usuario el radio del silo y lo guarda como número decimal en `r`.

3.  **`h = float(input(...))`**: Solicita la altura del silo y la guarda como número decimal en `h`.

4.  **`V = math.pi * (r ** 2) * h`**: Calcula el volumen `V` del silo usando la fórmula `π * radio² * altura`.

5.  **`cs = V * 0.75`**: Calcula la capacidad `cs` del silo en toneladas multiplicando el volumen por la densidad del maíz (0.75 ton/m³).

6.  **`print(...)`**: Muestra la capacidad calculada del silo en toneladas.
